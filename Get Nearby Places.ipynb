```python
from yelpapi import YelpAPI
import json

f = open('.api_keys')
api_key = json.loads(f.read())['yelp_api']
yelp_api = YelpAPI(api_key)


search_results = yelp_api.search_query(categories='grocery', longitude=-73.921460, latitude=41.595165, sort_by='best_match', limit=50)
```

```python
import pandas as pd
pd.DataFrame([x for x in search_results['businesses'] if x['categories'][0]['alias'] == 'grocery'])
```
